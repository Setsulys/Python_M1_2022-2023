```python
from itertools import *
```

## Exercice 1

```python
def queen_pbm(n):
    nbr=0;
    p = list(permutations(range(0,8)))   
    for i in p:
        p1 = {i-p[i] for i in range(len(p))}
        p2 = {i+p[i] for i in range(len(p))}
        if p1==p2==n:
            nbr+=1
    print(nbr)
            
```

```python
queen_pbm(4)
```

## Exercice 3

```python
def cornway(array,nbr):
    cpt =1
    lst=[]
    if(nbr == 0 ):
        return array
    for i in range(len(array)):
        if(i +1 < len(array)):
            if(array[i]==array[i+1]):
                cpt+=1
            else:
                lst.append(cpt)
                lst.append(array[i])
                cpt=1
        else:
            lst.append(cpt)
            lst.append(array[i])
    return cornway(lst,nbr-1)
```

```python
def all_cornway(start,nbr):
    for i in range(nbr):
        print("".join([str(x) for x in cornway(start,i)]))
```

```python
all_cornway([1],10)
```
